```python
dfq1= pd.read_csv("waqi-covid19-airqualitydata-2020Q1.csv",skiprows=4)
dfq2= pd.read_csv("waqi-covid19-airqualitydata-2020Q2.csv",skiprows=4)
dfq3= pd.read_csv("waqi-covid19-airqualitydata-2020Q3.csv",skiprows=4)
dfq4= pd.read_csv("waqi-covid19-airqualitydata-2020Q4.csv",skiprows=4)
```

```python
data= [dfq1,dfq2,dfq3,dfq4]
df2020 = pd.concat(data)
```

```python
df2020A=df2020[df2020["Specie"]=="so2"]
df2020A1=df2020A[df2020A["max"]>499]
df2020A1
```

```python
df2020A1["Country"].unique()
```

```python
df2020CT=df2020A1.groupby("City").mean("max")
df2020CT=df2020CT.reset_index()
```

```python
plt.figure(figsize=(20,5))
df2020CT.plot.bar(x= "City", y="max", rot=60)
plt.show()
```

```python
df2020A=df2020[df2020["Specie"]=="so2"]
df2020A2=df2020A[df2020A["max"]>230]
df2020A2
```

```python
df2020A2["Country"].unique()
```

```python
df2020COU=df2020A2.groupby("Country").mean(max)
df2020COU.reset_index()
df2020COU.sort_values(by=["max"], inplace= True, ascending= False)
```

```python
df2020COU=df2020COU.reset_index()
```

```python
df2020COU
```

```python
plt.figure(figsize=(20,5))
df2020COU.plot.bar(x= "Country", y="max", rot=60)
plt.show()
```
